```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\Dell\\Downloads\\FMCG_Assignment_IOTA\\Raw Sales Data.csv")

print(df.info())

df.head() 
```

```python
df['Qty']=df['Qty'].str.replace(",","")
df['Qty']=df['Qty'].astype(int)
df['Qty'].dtype
```

```python
df['Sales']=df['Sales'].str.replace(",","")
df['Sales']=df['Sales'].astype(int)
df['Qty'].dtype
```

```python
#1.Month Wise Qty Sold
df_Month_Qty=df.groupby(['Month'])['Qty'].sum()
df_Month_Qty=pd.DataFrame(df_Month_Qty.reset_index(0))
df_Month_Qty

```

```python
#2.Category wise Qty sold
df_Category_Qty=df.groupby('Category')['Qty'].sum()
df_Category_Qty=pd.DataFrame(df_Category_Qty.reset_index(0))
df_Category_Qty
```

```python
#3.Sub-channel wise total sales
df_Subchannel_Qty=df.groupby('Sub-Channels')['Sales'].sum()
df_Subchannel_Qty=pd.DataFrame(df_Subchannel_Qty.reset_index(0))
df_Subchannel_Qty
```

```python
#4.Highest sale in which month

df_high=df['Sales'].max()
df_high_sales= df[df['Sales']==df_high]

df_high_sales['Month'].values[0]
df_high_sales=df_high_sales['Month']
df_high_sales=pd.DataFrame(df_high_sales.reset_index(0))
df_high_sales
```

```python
#5.Highest selling product (based on sales) (packaging quantity doesn't matter. Ex. consider "GET SLIM JUICE 1 L" and "GET SLIM JUICE 500 ML" as one product)
df['Products']=df['Product Name'].str.extract('([A-Z+\s]+)')
df_high1=df.groupby('Products')['Sales'].sum()
Sort_pdt=df_high1.sort_values(ascending=False)
df_high_selling=pd.DataFrame(Sort_pdt.head(1).reset_index(0))
df_high_selling
```

```python
#List of different products company is offering (packaging quantity doesn't matter. Ex. consider "GET SLIM JUICE 1 L" and "GET SLIM JUICE 500 ML" as one product)
df_Category=df.groupby('Products')['Sales'].sum()
df_Category=pd.DataFrame(df_Category.reset_index(0))
df_Category
```

```python
writer = pd.ExcelWriter('FMCG_Summary.xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name='Data', index=False)
df_Month_Qty.to_excel(writer, sheet_name='df_Month_Qty', index=False)
df_Category_Qty.to_excel(writer, sheet_name='df_Category_Qty', index=False)
df_Subchannel_Qty.to_excel(writer, sheet_name='df_Subchannel_Qty', index=False)
df_high_sales.to_excel(writer, sheet_name='df_high_sales', index=False)
df_high_selling.to_excel(writer, sheet_name='df_high_selling', index=False)
df_Category.to_excel(writer, sheet_name='df_Category', index=False)
writer.save()
```
